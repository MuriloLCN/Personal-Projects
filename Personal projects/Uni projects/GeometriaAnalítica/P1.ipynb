## Exercício 1)

Heurística:

Objetivo: Determinar a área do triângulo ABC a partir de seus pontos médios MNP

Para calcular a área do triângulo ABC sabendo que MNP são seus pontos médios, basta observarmos que o triângulo formado 
entre seus pontos médios tem área igual a 1/4 da área do triângulo original (metade da base e metade da altura --> 1/4).

Obs: Isso pode ser provado com semelhança de triângulos e proporção entre seus lados, mas também montei uma visualização nesse link (https://www.geogebra.org/calculator/fdrbutds) que mostra que para qualquer alteração feita no triângulo (inclusive para dimensões superiores, contanto que ainda seja um triângulo), essa relação continua verdadeira - a linha vertical verde tem a equação x = (Área maior)/(Área média) = 4.

Obs 2: Utilizando isso juntamente com o baricentro, retas e outras propriedades, podemos achar as coordenadas dos pontos ABC, porém isso não foi necessário neste exercício.

Para calcularmos a área do triângulo MNP basta montarmos o produto vetorial dos vetores MN e MP, que nos dará a área do paralelogramo formado entre eles, e então basta vermos que a área do triângulo é a metade da área do paralelogramo formado.

Assim temos:

Área(ABC) = 4 * (Área MNP)

e

Área(MNP) = 1/2 * abs((MN x MP))   

```python
def vetorAPartirDePontos(pontoA: list, pontoB: list) -> list:
    # Rotina que forma um vetor a partir de dois pontos - para deixar mais limpo o código
    res: list = []
    
    for i, item in enumerate(pontoB):
        res.append(item - pontoA[i])
        
    return res

def produtoVetorial(vetorA: list, vetorB: list) -> list:
    # Rotina que calcula o produto vetorial entre dois vetores
    res = [None] * 3
    
    res[0] = (vetorA[1] * vetorB[2] - (vetorB[1] * vetorA[2]))
    res[1] = (vetorA[2] * vetorB[0] - (vetorB[2] * vetorA[0]))
    res[2] = (vetorA[0] * vetorB[1] - (vetorB[0] * vetorA[1]))
    
    return res

def moduloVetor(vetor: list) -> float:
    # Rotina que retorna o módulo (tamanho) de um vetor
    res = 0
    
    for item in vetor:
        res += (item * item)
        
    return sqrt(res)
    
M = [0, 1, 3]
N = [3, -2, 2]
P = [1, 0, 2]

MN = vetorAPartirDePontos(M, N)
MP = vetorAPartirDePontos(M, P)

prodVet_MN_MP = produtoVetorial(MN, MP)

modParalelogramo = moduloVetor(prodVet_MN_MP)

areaPontosMedios = modParalelogramo / 2

# Daria para simplificar o /2 e *4 com simplesmente *2, mas assim fica mais compreensível
areaTriangulo = 4 * areaPontosMedios

print(f"Área do triângulo ABC: {areaTriangulo}")
```

### Exercício 3)

Objetivo: Mostrar ao usuário todas as possibilidades diferentes de se pegar 3 vetores de uma lista de vários vetores de modo que esses 3 escolhidos sejam coplanares, i.e, todas as possibilidades de vetores coplanares no conjunto.

Heurística:

Primeiro, deverá ser pego os valores de entrada do usuário separados por vírgulas. Como o enunciado não especifica quantos vetores serão colocados no conjunto, todas as coordenadas de todos os vetores serão colocadas da forma: "x1, y1, z1, x2, y2, z2, x3, y3, z3, ...". A partir disso, podemos utilizar o .split() nas vírgulas e pegar os valores 3 à 3 para montar os vetores.

Em seguida, será montada uma lista vazia de subconjuntos coplanares, e então serão utilizados três laços for para checar cada permutação possível do conjunto de vetores e checar se o produto misto de cada permutação é nulo (i.e, coplanares), caso alguma permutação seja e ela ainda não esteja na lista de subconjuntos coplanares, ela será adicionada. Por fim, esta lista será mostrada ao usuário.

```python
def produtoVetorial(vetorA: list, vetorB: list) -> list:
    # Rotina que calcula o produto vetorial entre dois vetores
    res = [None] * 3
    
    res[0] = (vetorA[1] * vetorB[2] - (vetorB[1] * vetorA[2]))
    res[1] = (vetorA[2] * vetorB[0] - (vetorB[2] * vetorA[0]))
    res[2] = (vetorA[0] * vetorB[1] - (vetorB[0] * vetorA[1]))
    
    return res

def produtoEscalar(vetorA: list, vetorB: list) -> float:
    # Rotina que calcula o produto escalar entre dois vetores
    res = 0
    
    for i, valor in enumerate(vetorA):
        res += valor * vetorB[i]
    
    return res

def produtoMisto(vetorA: list, vetorB: list, vetorC: list) -> float:
    # Rotina que calcula o produto misto entre dois vetores
    return produtoEscalar(vetorA, produtoVetorial(vetorB, vetorC))

def subconjuntoCheck(vetores: list, conjunto: list) -> bool:
    # Checa se uma lista de 3 vetores está contida (em qualquer ordem) em um conjunto maior
    
    # Dá para utilizar um loop para isso mas como são apenas 6 possibilidades, acaba não fazendo muita diferença
    if ([vetores[0], vetores[1], vetores[2]] in conjunto):
        return True
    if ([vetores[0], vetores[2], vetores[1]] in conjunto):
        return True
    if ([vetores[1], vetores[0], vetores[2]] in conjunto):
        return True
    if ([vetores[1], vetores[2], vetores[0]] in conjunto):
        return True
    if ([vetores[2], vetores[1], vetores[0]] in conjunto):
        return True
    if ([vetores[2], vetores[0], vetores[1]] in conjunto):
        return True
    
    return False
    

coordString = input("Insira as coordenadas dos vetores (separadas com vírgulas): ")

# Trata a string para evitar erros - remove colchetes, espaços e parênteses
coordString = coordString.replace('[','')
coordString = coordString.replace(']','')
coordString = coordString.replace('(','')
coordString = coordString.replace(')','')
coordString = coordString.replace(' ','')

splt = coordString.split(',')

coords = [None] * len(splt)

vetores = []

if len(splt) % 3 != 0:
    print("A divisão do numero de coordenadas por 3 tem que ter resto zero")
    raise ValueError

for i, item in enumerate(splt):
    coords[i] = float(item)
    
# Começa em 0 e aumenta de 3 em 3 até o número de coordenadas inseridas    
for i in range(0, len(coords), 3):
    vet = [coords[i], coords[i + 1], coords[i + 2]]
    vetores.append(vet)

subconjuntos = []

# Para cada possível permutação entre os vetores 
for i in range(0, len(vetores), 1):
    
    # Obs: Começa no i
    for j in range(i, len(vetores), 1):
        
        # Se o vetor é igual ao anterior (repetido), pula
        if (vetores[j] == vetores[i]):
            continue
        
        # Obs: Começa no j
        for k in range(j, len(vetores), 1):
            
            # Se o vetor é igual a algum dos anteriores, pula
            if (vetores[k] == vetores[j] or vetores[k] == vetores[i]):
                continue
            
            # Se o produto misto é nulo
            if (produtoMisto(vetores[i], vetores[j], vetores[k]) == 0):
                # E se o subconjunto NÃO está na lista ainda
                if not subconjuntoCheck([vetores[i], vetores[j], vetores[k]], subconjuntos):
                    # Coloca eles na lista
                    subconjuntos.append([vetores[i], vetores[j], vetores[k]])
                
print(f"O numero de subconjuntos de 3 vetores que são coplanares: {len(subconjuntos)}")

for conj in subconjuntos:
    print(f"O conjunto {conj} é coplanar")
```
